```python
import numpy as np

from pybatteryid.inputdesign import generate_current_profile, generate_temperature_profile
from pybatteryid.plotter import plot_time_vs_current, plot_time_vs_temperature
```

### Example 5 - Identification Input Design

This file contains an example to generate an input current profile as well as a temperature profile for battery model identification.

Please see [1] for the parameters associated with the proposed signal design.

#### References

[1]

#### A. Generate current profile

```python
current_profile = generate_current_profile(mean_pulse_period=64, std_pulse_period=32,
                                           mean_rest_period=16, std_rest_period=8,
                                           amplitude_plus=1, amplitude_minus=0.1,
                                           std_random_amplitude=0.75,
                                           initial_amplitude=2, dc_offset=-1,
                                           no_of_pulses=300,
                                           upper_safety_limit=3, lower_safety_limit=-5)
```

```python
# Assuming 1 second as sampling interval
time = np.arange(len(current_profile)) / 3600
#
plot_time_vs_current([(time, current_profile)], units=('h', 'A'), colors=['darkblue'])
```

#### B. Generate temperature profile

```python
temperature_profile = generate_temperature_profile(positive_phase_length=3500, negative_phase_length=4500,
                                                   pulse_shift=5/6,
                                                   pulse_amplitude=21.25,
                                                   dc_offset=19.5, no_of_pulses=10)
```

```python
# Assuming 1 second as sampling interval
time = np.arange(len(temperature_profile)) / 3600
#
plot_time_vs_temperature([(time, temperature_profile)], units=('h', 'degC'), colors=['darkblue'])
```
