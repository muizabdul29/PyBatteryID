```python
import numpy as np

from pybatteryid.inputdesign import InputDesign
from pybatteryid.plotter import plot_time_vs_current
```

### Example 5 - Identification Input Design

This file contains an example to generate an input current signal for battery model identification.

#### A. Define parameters

Please see [X] for the parameters associated with the proposed signal design.

```python
inputdesign = InputDesign(mean_pulse_period=64, std_pulse_period=32,
                          mean_rest_period=16, std_rest_period=8,
                          amplitude_plus=1, amplitude_minus=0.1,
                          std_random_amplitude=0.75,
                          initial_amplitude=2, dc_offset=-1,
                          no_of_pulses=300,
                          upper_safety_limit=3, lower_safety_limit=-5)
```

### B. Generate signal profile

```python
current_profile = inputdesign.generate_signal_profile()
```

```python
# Assuming 1 second as sampling interval
time = np.arange(len(current_profile)) / 3600
#
plot_time_vs_current([(time, current_profile)], units=('h', 'A'), colors=['darkblue'])
```
